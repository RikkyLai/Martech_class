```python
# 分析basket 电影分类中的频繁项集和关联规则
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

# 数据加载
market = pd.read_csv('./Market_Basket_Optimisation.csv', header=None)
```

```python
market.head()
```

```python
transactions = []
col = set()
for i in range(0, 7501):
    t = []
    for j in range(0, 20):     
        t.append(str(market.values[i,j]))
        col.add(str(market.values[i,j]))
    transactions.append(t)
```

```python
col.remove('nan')
```

```python
len(col)
```

```python
transaction_f = []
col = list(col)
for i in range(len(transactions)):
    d = [0]*len(col)
    for j in range(len(transactions[i])):
        item = transactions[i][j]
        if item in col:
            d[col.index(item)] = 1
    transaction_f.append(d)
    
```

```python
data = pd.DataFrame(transaction_f, columns=list(col))
```

```python
data
```

```python
data.to_csv("Market_Basket1.csv")
```

```python
# 挖掘频繁项集，最小支持度为0.02
itemsets = apriori(data, use_colnames=True, min_support=0.003)
# 按照支持度从大到小进行时候粗
itemsets = itemsets.sort_values(by="support" , ascending=False) 
print('-'*20, '频繁项集', '-'*20)
print(itemsets)
# 根据频繁项集计算关联规则，设置最小提升度为2
rules =  association_rules(itemsets, metric='lift', min_threshold=3)
# 按照提升度从大到小进行排序
rules = rules.sort_values(by="lift" , ascending=False) 
#rules.to_csv('./rules.csv')
print('-'*20, '关联规则', '-'*20)
print(rules)
```
